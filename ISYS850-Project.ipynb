```python
import pandas as pd
import numpy as np

hours = pd.read_csv("annual_hours.csv")
happiness = pd.read_csv("happiness_2015.csv")
alcohol = pd.read_csv("drinks_2015.csv")
```

## Hours

```python
hours.head()
```

```python
country_hours2015 = hours.loc[hours['Year']==2015].groupby('Entity')['Average annual working hours per worker'].max()
country_hours2015.head()
```

```python
country_hours2015 = pd.DataFrame(country_hours2015)
country_hours2015
```

```python
country_hours2015['country'] = country_hours2015.index
country_hours2015 = country_hours2015[['country', 'Average annual working hours per worker']]
country_hours2015
```

## Happiness

```python
happiness
```

```python
happiness.columns
```

```python
bad_cols = ['Region','Standard Error', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 
            'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']
happiness.drop(bad_cols, axis=1, inplace=True)
happiness.sort_values(by=['Country'], inplace=True)
```

```python
happiness.reset_index(inplace=True)
del happiness['index']
```

```python
happiness
```

## Alcohol

```python
alcohol
```

```python
country_hours2015.columns = country_hours2015.columns.str.lower()
happiness.columns = happiness.columns.str.lower()
alcohol.columns = alcohol.columns.str.lower()

country_hours2015['country'] = country_hours2015['country'].str.lower()
happiness['country'] = happiness['country'].str.lower()
alcohol['country'] = alcohol['country'].str.lower()
```

```python
country_hours2015.index = range(len(country_hours2015.index))
country_hours2015.head()
```

```python
happiness.head()
```

```python
alcohol.head()
```

```python
hours_happiness = country_hours2015.merge(happiness, left_on = 'country', right_on = 'country')
hours_happiness
```

```python
final_df = hours_happiness.merge(alcohol, left_on = 'country', right_on = 'country')
final_df
```

```python
final_df.to_csv('hours_happiness_alcohol.csv')
```
